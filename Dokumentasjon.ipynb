### **Metode 2:** Copy-paste av tabellen fra nettsiden

Denne viste seg å være ganske effektiv og enkel å utføre. Aller først kopierer man tabellen fra nettsiden over til notepad for å få følgende .txt fil:
   
    1	1	1
    2	1	1, 2
    3	2	1, 3
    4	2	1, 2, 4
    5	4	1, 5

... og så videre. Inne på notepad kan man erstatte ", " (komma-mellomrom) med "," (komma) slik at listene av faktorer i den tredje kolonnen blir én sammenhengende rekke med tegn. Man får dermed følgende .txt fil: 

https://raw.githubusercontent.com/wolneberg/Eulers-totientfunksjon/master/Skriving%20av%20csv%20fil/textfil.txt

Denne kan så importeres hit med urrlib.request og pylab

```python
import urllib.request    # For å bruke urllib.request.urlopen()

tekst = urllib.request.urlopen("https://raw.githubusercontent.com/wolneberg/Eulers-totientfunksjon/master/Skriving%20av%20csv%20fil/textfil.txt")


import pylab as pl    # For å bruke pl.loadtxt()

data = pl.loadtxt(tekst,  str)    # "data" blir en array med 3 kolonner der elementene er strenger

print(data)
```

Objektet "data" er nå en array med strenger. Målet er at elementene i første og andre kolonne, altså n og phy(n), blir integers, mens de i tredje kolonne, altså faktorene, blir lister av integers. Dette kan gjøres med følgende kode:

```python
første = []    # Første kolonne
andre = []    # Andre kolonne
tredje=[]    # Tredje kolonne

for x in data[:,0]:          # Første kolonne blir heltallene n
    første.append(int(x))

for x in data[:,1]:         # Andre kolonne blir heltallene phy(n)
    andre.append(int(x))

for x in data[:,2]:     # Tredje kolonne blir en liste av heltallene som er faktorer av n
    x = x.split(",")
    x = [int(i) for i in x]
    tredje.append(x)

tabellen = []    # den endelige tabellen, som egentlig blir en liste av lister

for x in range(0,500):           # Hver rad i "tabellen" blir fylt inn
    rad = [første[x], andre[x], tredje[x]]
    tabellen.append(rad)
    
for x in tabellen:
    print(x)
```

Nå har vi en veldig oversiktlig tabell. Neste steg er å omdanne dette til en .csv fil. Dette kan gjøres med pandas:

```python
import pandas as pd

columns = ["n","phy(n)","faktorer"]    # Kolonnene i tabellen 

dataframe = pd.DataFrame(tabellen, columns = columns)    # "tabellen" gjøres om til en pandas dataframe

dataframe.to_csv("Euler500verdier.csv")
```

Nå har vi filen "Euler500verdier.csv" lagret i samme mappe som denne jupyter filen. Videre kan vi analysere denne.

### **Metode 3:** Funksjon som beregner phy(n) ved brute force og returnerer en .csv fil

Siden Eulers Totientfunksjon er en matematisk funksjon, så kan man lage et program som regner ut phy(n) ved brute force. Dermed kan man også programmere en funksjon som kan lage en .csv fil lik den som man kan lage med metode 2. Funksjonen er som følger:

```python
def TotientGenerator(maxverdi):     # "maxverdi" er hvor mange rader .csv filen inneholder, altså hvor mange naturlige tall fra 1 til "maxverdi" som inkluderes
    from pandas import DataFrame

    nliste = [1]      # De tre kolonnene starter som lister, og inneholder verdiene for n=1
    phyliste = [1]
    faktorliste = [[1]]

    for x in range(2,maxverdi+1):    # Fra n=2 til og med "maxverdi" blir kolonnene forlenget 
        n = x
        phy = 0
        faktorer = []
        for y in range(1,x+1):    # Faktorene til n beregnes og legges til i listen av faktorer
            if x%y == 0:
                faktorer.append(y)

        for y in faktorliste:    # Med brute force sjekker programmet hvor mange tall mindre enn n som er innbyrdes primske med n
            innbyrdes = True
            for z in range(1,len(faktorer)):
                if faktorer[z] in y:
                    innbyrdes = False
            if innbyrdes == True:
                phy += 1

        nliste.append(n)        # Verdiene legges til sine respektive kolonner
        phyliste.append(phy)
        faktorliste.append(faktorer)

    tabell = []                    # Den endelige tabellen ferdigstilles 
    for x in range(0,maxverdi):
        rad = [nliste[x],phyliste[x],faktorliste[x]]
        tabell.append(rad)
        
    kolonner = ["n","phy(n)","faktorer"]
    filnavn = "Totient"+str(maxverdi)+"verdier.csv"         # "filnavn" er skrevet slik at navnet på .csv filen reflekterer hvor mange verdier av n er inkludert
    dataframe = pd.DataFrame(tabell, columns = kolonner)
    
    dataframe.to_csv(filnavn)    # .csv filen lagres i samme mappe som denne jupyter filen, funksjonen behøver ikke returnere noe
```

Nå kan vi kalle på den funksjonen og få en .csv fil med så mange verdier vil: 1000, 5000, osv. Dermed vil vi ha et større datasett å analysere.

Hvis vi kaller på funksjonen med 500 som parameter, så burde vi få en .csv fil som er prikk lik den filen vi lagde i metode 2, som også har 500 verdier av n. Vi kan dermed kontrollere om funksjonen funker som den skal.

```python
TotientGenerator(500)    # en .csv fil med 500 verdier lagres i mappen gjennom funksjonen
```

Nå har to .csv filer lagret i samme mappe som denne jupyter filen. "Euler500verdier.csv" er laget utifra nettsiden, og "Totient500verdier.csv" som er generert uten hjelpemidler. Vi kontrollerer likheten:

```python
import pandas as pd

generert = pd.read_csv("Totient500verdier.csv")    # Den genererte .csv filen leses
ekstrahert = pd.read_csv("Euler500verdier.csv")    # Den ekstraherte .csv filen leses

ulikheter = 0

for x in range(0,500):
    for y in range(0,3):
        if generert.iloc[x,y] != ekstrahert.iloc[x,y]:
            ulikheter += 1

print("Det er",ulikheter,"ulikheter i tabellene")
            
```
